```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loss(yhat, y): 
    """
    yhat: ndarray -> vector
    y:    ndarray -> vector
    """
    return - (y * np.log(yhat) + (1 - y) * np.log( 1 - yhat))
```

```python
x = np.arange(100)
delta = np.random.uniform(-10,10, size=(100,))
y_line = .4 * x +3 + delta
```

```python
plt.plot(x,y_line)
plt.show()
```
